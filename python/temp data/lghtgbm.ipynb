```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import os
```

```python
train_data= pd.read_excel("all_past_year_paramater_include_yeild.xlsx")
train_data.drop(columns=["Unnamed: 0", "Value (bu)"],inplace=True)
```

```python
data_yield = pd.read_csv("WheatYield_reset.csv")
#將字串轉成浮點數
for j in range(len(data_yield["Value (bu)"])):
    n = data_yield["Value (bu)"][j].split(",")
    number = ''
    for k in n:
        number += k
        data_yield["Value (bu)"][j] = float(number)
```

```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split


train_data = pd.get_dummies(train_data, columns=['Encode'])
# 將特徵和目標變數分開
```

```python
X = train_data
y = data_yield["Value (bu)"].astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
# 建立 LightGBM 資料集
train_data = lgb.Dataset(X_train, label=y_train)
```

```python
# Creating a Function for MAPE
import numpy as np

def mape(test, pred):
    test, pred = np.array(test), np.array(pred)
    mape = np.mean(np.abs((test - pred) / test))
    return mape
```

```python
ratio = 100
ratiovalues = [i for i in range(10, ratio, 10)]
depth = 21
depthvalues = [i for i in range(1, depth)]
leaf = 10
leafvalues = [i for i in range(1, leaf)]
```

```python
X_test
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
# 定義參數空間
param_grid = {
    'num_leaves': leafvalues,
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': depthvalues,
    'min_child_samples': [10, 20, 30]
}

# 設定 LightGBM 參數
params = {
    'objective': 'regression',
    'metric': 'rmse'
}

# 使用網格搜尋進行調參
grid_search = GridSearchCV(estimator = lgb.LGBMRegressor(**params), param_grid = param_grid, cv = 3)
grid_search.fit(X_train, y_train)

# 取得最佳參數組合
best_params = grid_search.best_params_

# 使用最佳參數重新訓練模型
model = lgb.LGBMRegressor(**params, **best_params)
model.fit(X_train, y_train)

# 預測測試集
y_pred = model.predict(X_test)

# 計算均方根誤差 (Root Mean Squared Error, RMSE)
rmse = mean_squared_error(y_test, y_pred, squared = False)

mape_r = mape(y_test,y_pred)

print('Best parameters:', best_params)
print('RMSE:', rmse)
print('MAPE:', mape_r)
r2 = 1 - ((y_pred - y_test) ** 2).sum() / ((y_test - y_test.mean()) ** 2).sum()

print('R-squared:', r2)
```

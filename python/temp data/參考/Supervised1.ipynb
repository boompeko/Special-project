## Classification

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris)
```

```python
from sklearn import tree  # Tree
from sklearn.datasets import load_iris # Iris
iris = load_iris() # Iris dataset
X = iris.data  # Features
y = iris.target  # Types
clf = tree.DecisionTreeClassifier() # Descision Tree Function
clf.fit(X,y) # Training Module
```

```python
tree.plot_tree(clf) # Draw Tree
```

```python
import graphviz  # graphviz

dot_data = tree.export_graphviz(clf, out_file=None, filled=True, rounded=True, leaves_parallel=True)
# Build data
graph = graphviz.Source(dot_data) # Source for Figure
graph # Draw Tree
```

```python
# Decistion Tree
from sklearn import tree
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = tree.DecisionTreeClassifier()
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = RandomForestClassifier()  #Default n_estimators=100
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
#XGBoost 1 cannot be executed in new verion
from xgboost.sklearn import XGBClassifier
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = XGBClassifier()
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
#XGBoost 2 can be executed in all versions
from xgboost.sklearn import XGBClassifier
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
X = iris.data
y = iris.target
clf = XGBClassifier()
clf.fit(X,y)
target = np.array([[21,10,5,10]])
y_pred = clf.predict(target)
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
# Decistion Tree
from sklearn import tree
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = tree.DecisionTreeClassifier()
clf.fit(X,y)
y_pred = clf.predict([[4.8, 6.2, 2.3, 8.1]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = RandomForestClassifier()
clf.fit(X,y)
y_pred = clf.predict([[4.8, 6.2, 2.3, 8.1]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
#XGBoost 2
from xgboost.sklearn import XGBClassifier
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
X = iris.data
y = iris.target
clf = XGBClassifier()
clf.fit(X,y)
target = np.array([[4.8, 6.2, 2.3, 8.1]])
y_pred = clf.predict(target)
print(y_pred)
print(iris.target_names[y_pred[0]])
```

## Other Classification Modules

```python
# SVM
from sklearn import svm
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = svm.SVC()
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
# KNN
from sklearn import neighbors
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = neighbors.KNeighborsClassifier()
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
clf = LogisticRegression()
clf.fit(X,y)
y_pred = clf.predict([[21,10,5,10]])
print(y_pred)
print(iris.target_names[y_pred[0]])
```

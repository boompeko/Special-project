```python
tran=pd.concat([df['Average Temperature_x'],df['encode_state']],axis=1)
```

```python
# 假设你有一个名为 df 的 DataFrame 对象，其中包含你的时间序列数据和虚拟变量
# 时间序列数据存储在列 'TS1' 和 'TS2' 中，虚拟变量存储在列 'Season' 中

# 将 'Season' 列转换为虚拟变量
##df_with_dummies = pd.get_dummies(df, columns=['encode_state']), drop_first=True)
df_with_dummies = pd.get_dummies(tran, columns=['encode_state'])
```

```python
ts_data = df_with_dummies[['Average Temperature_x']] ## 時間序列
exog_data = df_with_dummies.drop(['Average Temperature_x'], axis=1) ## 外生變數
```

```python
import numpy as np
## 轉為數列
ts_data_array = np.asarray(ts_data) 
exog_data_array = np.asarray(exog_data)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python

import statsmodels.api as sm
```

```python
# 擬合 ARIMAX 模型
model = sm.tsa.ARIMA(endog=ts_data_array, exog=exog_data_array, order=(11, 0, 10))
model_fit = model.fit()
```
